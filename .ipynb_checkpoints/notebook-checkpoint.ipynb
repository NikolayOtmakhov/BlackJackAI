```python
import deck
import player
import pandas as pd

# create deck and player
d = deck.Deck()
pl = player.Player()


class mem:
    count = 0
    aces = 0
    used_aces = 0
    value = 0
    result = 0
    save_value = 0
    cards = 0


def give_player_card_and_count():
    card = d.take_card()
    value = card.value
    mem.value = pl.card_value_total()
    mem.cards += 1
    if value > 9:
        mem.count -= 1
    if value < 7:
        mem.count += 1
    if value == 11:
        mem.aces += 1
    pl.give_card(card)

df = pd.DataFrame(columns = ["Count", "Aces", "Aces_Used", "Cards", "Value" , "Result"] )
# df = pd.read_csv("file1.csv", index_col=0)


data_entries = 0
while data_entries < 500000:
    give_player_card_and_count()
    give_player_card_and_count()
    while mem.value < 22:
        if data_entries % 100 == 0:
            print(data_entries)
        if data_entries % 50000 == 0:
            df.to_csv("file1.csv")
            print(f"DF SIZE: {df.size}")
        mem.result = 0
        mem.save_value = mem.value
        give_player_card_and_count()
        if mem.value < 22 and mem.value >= mem.save_value:
            mem.result = 1
#         print(f"{mem.value} {mem.save_value}")
        if mem.value < 22 and mem.value < mem.save_value:
            mem.used_aces += 1
            result_temp = 0.5
            if mem.save_value == 21:
                result_temp = 0
            elif mem.save_value < 14:
                result_temp = 1
            else:
                result_temp = 0.5 - (mem.save_value - mem.value)*0.1
                if result_temp < 0:
                    result_temp = 0
                if mem.used_aces < mem.aces:
                    result_temp = 0.5
            mem.result = result_temp
        df.loc[len(df)] =([mem.count, mem.aces, mem.used_aces, mem.cards, mem.save_value , mem.result])
        data_entries += 1
        if d.cards_left() < 20:
            d = deck.Deck()
            mem.count = 0
    mem.value = 0
    mem.aces = 0
    mem.result = 0
    mem.cards = 0
    mem.used_aces = 0
    pl.init_or_reset_hand()
    

```

```python
import tensorflow as tf
import keras
from keras import layers
from keras.layers.experimental import preprocessing
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split


df = pd.read_csv("file1.csv", index_col=0)
df = sklearn.utils.shuffle(df)
df_features = df
df_label = df_features.pop("Result")

df_features = np.array(df_features)

X_train, X_test, y_train, y_test = train_test_split(
    df_features, df_label, test_size=0.33)

normalize = preprocessing.Normalization()
normalize.adapt(df_features)

blackjack_model = tf.keras.Sequential([
    normalize,
    layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.1),
    layers.Dense(64),
    layers.Dense(1)
])

blackjack_model.compile(loss = tf.losses.MeanSquaredError(),
                      optimizer = tf.optimizers.Adam())

blackjack_model.fit(X_train, y_train, epochs=10, 
    validation_data=(X_test, y_test))

np.round(blackjack_model.predict([[2,0,3,13]]))[0][0]

blackjack_model.save('saved_model/blackjack_model')
```
